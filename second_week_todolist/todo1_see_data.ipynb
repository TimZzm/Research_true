```python
import dedalus.public as d3
import logging

logger = logging.getLogger(__name__)
import copy
import h5py
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
from dedalus.extras import plot_tools
import os
from os import listdir

matplotlib.use('Agg')


folder_dir = "snapshots"

file_paths = [os.path.join(folder_dir, file) for file in listdir(folder_dir) if
              os.path.isfile(os.path.join(folder_dir, file)) and file.endswith('.h5')]
file_paths.sort()
print(file_paths)

for file in file_paths:
    with h5py.File(file, mode='r') as file:
        ls = list(file.keys())
        print(ls)
        values = list(file.values())
        print(values)
        scales_in_h5 = file.get('scales')
        scales_array = np.array(scales_in_h5)
        tasks_in_h5 = file.get('tasks')
        tasks_array = np.array(tasks_in_h5)
        print(scales_array)
        for each_scale in scales_array:
            print(f'\n\n{each_scale}:')
            print(np.array(scales_in_h5[each_scale]))
            print('\n\n')
        print(tasks_array)
        for each_task in tasks_array:
            print(f'\n\n{each_task}:')
            print(np.array(tasks_in_h5[each_task]))
            print('\n\n')




```
